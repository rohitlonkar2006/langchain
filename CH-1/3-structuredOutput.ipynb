### Guiding in prompts

```python
import os
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model = "qwen/qwen3-32b")

response = llm.invoke("Tell me a joke generate the output in key-value pair format with the following key: setup,punchline")
response.content
```

### Structured output using pydentic models

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    setup: str = Field(description= "The setup for the capital of india")
    punchline: str = Field(description="the punchline for the capital of india")
      
llm_structured_op = llm.with_structured_output(llm_schema)
llm_structured_op.invoke("tell me a capital of india")   
```

### Using TypedDict

```python
from typing_extensions import TypedDict, Annotated

class llm_schema_td(TypedDict):
    setup:str
    punchline:str

llm_with_structured_op = llm.with_structured_output(llm_schema_td)
llm_with_structured_op.invoke("tell me a joke")
```
