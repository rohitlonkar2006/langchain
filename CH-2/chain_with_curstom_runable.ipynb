```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

### Chain With Custom Runable

```python
# Task 1[prompt template]
prompt_template = ChatPromptTemplate.from_messages([
    ("system","you are a helpfull assistant"),
    ("human","{input}")    
])
```

```python
# Task 2 [llm model]
llm = ChatGroq(model = "qwen/qwen3-32b", temperature = 0)
```

```python
# Task 3 [parser]
str_parser = StrOutputParser()
```

```python
# Task 4 [custom runable]
from langchain_core.runnables import RunnableLambda 

def dictionary_maker(text: str) ->dict:
    return {"text" : text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)   
```

```python
#Task 5 [Template for post]

prompt_post = ChatPromptTemplate.from_messages([
    ("system","You are a social media post generator. "),
    ("human", "Ctrate a post for the following text for the linkedin: {text}")
])
```

```python
#Task 6[Generate the post]

llm = ChatGroq(model = "qwen/qwen3-32b")
```

```python
#Task 7 [output prser]

str_parser = StrOutputParser()
```

```python
# CHAIN

chain = prompt_template | llm | str_parser | dictionary_maker_runnable | prompt_post | llm | str_parser

chain.invoke("what is the capital of france?")
```
